```python
#Feature Selection

selected_columns = ['Age', 'Gender', 'Air Pollution', 'Genetic Risk', 'chronic Lung Disease',
                    'Smoking', 'Passive Smoker', 'Shortness of Breath', 'Level']
selected_data = df.loc[:, selected_columns]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
Le = LabelEncoder()
selected_data['Level'] = Le.fit_transform(selected_data['Level'])
```

```python
#Training the dataset

X = selected_data.iloc[:, :-1]
Y = selected_data.iloc[:, -1]
train_X, test_X, train_Y, test_Y = train_test_split(X,Y, train_size=0.5, random_state=1)

#train_size gives size of the training set
#The random state hyperparameter in the train_test_split() function controls the shuffling process.
```

```python
from sklearn.preprocessing import StandardScaler
Sc = StandardScaler()
train_X = Sc.fit_transform(train_X)
test_X = Sc.transform(test_X)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(train_X, train_Y)
prediction = LR.predict(test_X)
from sklearn.metrics import accuracy_score
print("The accuracy score of the model is ", accuracy_score(test_Y, prediction))
```
